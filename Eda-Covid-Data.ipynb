```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("covid_19_india.csv")
df
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='State/UnionTerritory',data=df)
plt.title("Total Cases")
plt.show()
```

```python
sns.displot(df.Confirmed)
plt.title("Cases")
plt.show()
```

```python
plt.figure(figsize=(20,10))
tags=df['State/UnionTerritory'].value_counts()
tags.plot(kind='pie',autopct='%1.1f%%',label='')
plt.show()
```

```python
df[['Cured','Deaths']].hist(bins=10,figsize=(10,5),color='red')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title('CURED VS DEATH')
sns.scatterplot(x=df['Cured'],y=df['Deaths'])
plt.xlabel('Cured')
plt.ylabel('Deaths')
plt.show() 
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation between columns')
plt.show()
```
